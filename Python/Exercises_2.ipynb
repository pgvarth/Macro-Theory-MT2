### Άσκηση 3

Συντελεστής Gini είναι μια μέτρηση της στατιστικής διασποράς που προορίζεται να αντιπροσωπεύει την κατανομή του εισοδήματος ή του πλούτου των κατοίκων ενός έθνους και είναι το πιο συχνά χρησιμοποιούμενο μέτρο της ανισότητας.

Κατασκευάστε μια συνάρτηση στην Python που να δέχετε ως input μία λίστα από εισοδήματα (σε δολλάρια άνα έτος) και δείχνει τον συντελεστή Gini.

$$G = \left( \frac{2 \sum_{i=1}^{n} i \cdot y_i}{n \sum_{i=1}^{n} y_i} \right) - \frac{n + 1}{n} $$

όπου:

- G είναι ο συντελεστής Gini (που κυμαίνεται μεταξύ 0 και 1 ως κλάσμα)

- n είναι ο αριθμός των εισοδημάτων (άρα των παρατηρήσεων που έχουμε)

- $y_i$ είναι το εισόδημα του νοικοκυριού $i$

- $i$ είναι η θέση στην αύξουσα κατάταξη του εισοδήματος 

Οπότε ουσιαστικά ο συντελεστής Gini είναι ένας **σταθμισμένος** μέσος των εισοδημάτων.

Για να είμαστε πιο συγκεκριμένοι, ας πούμε ότι η λίστα των εισοδημάτων που θέλουμε να ελέγξουμε είναι:

[10000, 20000, 25000, 40000, 60000]

Hint: Γι αυτές τις τιμές, ο συντελεστής Gini είναι 0.30967741935483883.

Επιπλέον ερώτημα: Τι θα συμβεί στον συντελεστή Gini αν αυξήσουμε το εισόδημα του μεγαλύτερου στην κατάταξη εισοδήματος; Φερ' ειπείν, αν αυξήσουμε το 60.000 σε 100.000.
